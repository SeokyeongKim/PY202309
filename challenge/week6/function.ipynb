```python
import random

def get_number():
    return random.randrange(1, 46)

def set_lotto_numbers():
    lotto = []
    while len(lotto) < 6:
        num = get_number()
        if num not in lotto:
            lotto.append(num)
    return lotto

def match_numbers(user_inputs, lotto_numbers):
    return list(set(user_inputs).intersection(lotto_numbers))

def check_prize(same_numbers):
    num_matching = len(same_numbers)
    if num_matching == 6:
        print("Congratulations! You've won a 1st prize")
    elif num_matching == 5:
        print("You've matched 5 numbers! You've won a 2nd prize.")
    elif num_matching == 4:
        print("You've matched 4 numbers! You've won a 3rd prize.")
    elif num_matching == 3:
        print("You've matched 3 numbers! You've won a 4th prize.")
    else:
        print("No prize :(")

user_numbers = []
lotto = []

print("** Start Lotto **")
print("1) Set the lotto numbers")

# set the lotto numbers
lotto = set_lotto_numbers()
#print(lotto)
print("2) Set the user inputs")

# get 6 numbers from the user
while len(user_numbers) < 6:
    try:
        user_input = int(input("Enter %d-number: " % (len(user_numbers) + 1)))
        if 1 <= user_input <= 45:
            if user_input not in user_numbers:
                user_numbers.append(user_input)
            else:
                print("Same number")
        else:
            print("No valid number")
    except:
        break

print("3) Match the numbers")

# match the numbers
matching_numbers = match_numbers(user_numbers, lotto)

print("4) Check the prize")

# check for the prize
check_prize(matching_numbers)

# print the user's numbers
print("5) Sort the lists")
print("=== Summary ===")

# sort the lotto numbers
lotto.sort()
print("Lotto Numbers: ", end=' ')
for number in lotto:
    print(number, end=' ')

print("\nUser Numbers: ", end=' ')
for number in user_numbers:
    print(number, end=' ')
```
